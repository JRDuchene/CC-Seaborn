```python
import seaborn as sns
mpg = sns.load_dataset("mpg")
mpg
```

```python
sns.lineplot(data=mpg, x="horsepower", y="mpg")
```

```python
sns.catplot(data=mpg, x="origin", y="mpg", kind="bar")
```

```python
sns.lineplot(data=mpg, x="model_year", y="mpg")
```

```python
sns.catplot(data=mpg, x="cylinders", y="mpg", kind="bar")
```

```python
sns.relplot(data=mpg, x="displacement", y="mpg", hue="origin")
```

```python
#Findings:
1. In general, as horsepower increases fuel economy tends to declin
2. Fuel economy has improved over time as technology improves
3. Automobiles from Japan tend to have greater fuel economy while also having less displacement
4. Four and Five cylinder vehicles tend to have the greatest fuel economy, while there is far more variance in the 5 cylinder vehicle.
```
