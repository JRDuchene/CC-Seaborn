```python
#Create a visual showing the relationship between horsepower and fuel economy.
import seaborn as sns
mpg = sns.load_dataset("mpg")
sns.lineplot(data=mpg, x="horsepower", y="mpg")
```

```python
#Create a table summarizing fuel economy by country of origin
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
mpg = sns.load_dataset("mpg")
sns.scatterplot(data=mpg, x='origin', y='mpg')
data = mpg.groupby('origin')['mpg'].mean().round(2).reset_index().values.tolist()
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=data, colLabels=['Origin', 'MPG'], loc='center', cellLoc='center')
plt.show
```

```python
#Create a visual showing how fuel economy has changed over time
sns.lineplot(data=mpg, x="model_year", y="mpg")
```

```python
#Create two other visuals or tables related to fuel economy.
sns.catplot(data=mpg, x="cylinders", y="mpg", kind="bar")
sns.relplot(data=mpg, x="displacement", y="mpg", hue="origin")
```

```python
#Findings
#1. In general, as horsepower increases fuel economy tends to decline
#2. Fuel economy has improved over time as technology improves
#3. Automobiles from Japan tend to have greater fuel economy while also having less displacement
#4. Four and Five-cylinder vehicles tend to have the greatest fuel economy, while there is far more variance in the 5 cylinder vehicle.
```
